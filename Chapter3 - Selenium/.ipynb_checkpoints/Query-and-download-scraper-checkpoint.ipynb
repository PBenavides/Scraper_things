```python
dict_tablas_links_noticias = {}
def scrap_name_sub_links(soup, dict_tablas_links_noticias):
    soup = BeautifulSoup(html, "html.parser")
    main_table_info = soup.find('table',{'id':'MainContent_grdHechosImportancia2'}) #Estoy dentro de la tabla principal
    for tabla_noticia in main_table_info.find_all('tr'):
        #Para cada tabla de noticia.
        #---------EL NOMBRE ES NULO SI LA EMPRESA SE REPITE. EL RESULTADO FINAL ES EL QUE TIENE NOMBRE NO NULO--------------------- 
        soup_obj_nombre = tabla_noticia.find('span',{'class':'titulo-empresa-azul'})
        try:
            nombre_empresa = soup_obj_nombre.text 
            links_obj_tabla = [link['href'] for link in tabla_noticia.find_all('a', href = re.compile(r'(http://www.smv.gob.pe/ConsultasP8/documento.aspx?)'))]
            subtitulo_noticias = [subtitle.text for subtitle in tabla_noticia.find_all('h5',{'class':'card-title'})]
            dict_tablas_links_noticias[nombre_empresa] = {'links': links_obj_tabla,'subtitulos':subtitulo_noticias}
        except AttributeError:
        #Si el nombre es nulo, se ignorará el caso y se seguirá con el código. 
            pass
    return dict_tablas_links_noticias
```
